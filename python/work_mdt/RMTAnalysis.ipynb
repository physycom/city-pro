```python
from RMTAnalysis import *
import time
import os
PlotDir = os.getcwd()
Sources = 1000
TimeIntervals = 96
Signal = np.random.rand(Sources,TimeIntervals)

print("Intialize")
t0 = time.time()
RMA = AnalysisCorrelationRandomMatrix(Signal,PlotDir)
print("AnalysisCorrelationRandomMatrix: ",time.time()-t0)
# NOTE: Returns Also The Correlation Matrix
t0 = time.time()
RMA.ComputeCorrelationMatrix()
print("ComputeCorrelationMatrix: ",time.time()-t0)
# Expansive Computations
#    RMA.ComputeEigenvalues()
t0 = time.time()
RMA.ComputeEigenvectorsAndEigenValues()
print("ComputeEigenvectors: ",time.time()-t0)
t0 = time.time()
RMA.SortEigvalsEigVec()
t1 = time.time()
print("SortEigvalsEigVec: ",t1-t0)
t0 = time.time()
RMA.SplitRealAndImaginary()
t1 = time.time()
print("Split Real-Imag:",t1-t0)
#
t0 = time.time()
RMA.ComputeDyadicExpansion()
print("DyadicExpansion: ",time.time()-t0)
t0 = time.time()
RMA.MarchenkoPasturLimits()
print("MarchenkoPasturLimits: ",time.time()-t0)
RMA.PlotAnalysis()

RMA.ComputeDistributionLambda()
```
